```python
import numpy as np
from scipy.spatial.distance import hamming
```

```python
motifs = []
```

```python
with open('subtle_motif_dataset.txt') as f:
    for line in f:
        motifs.append(line.strip())
```

```python
nucleotidos = ['A', 'C', 'T', 'G']

for i in range(len(motifs)):
    motifs[i] = ''.join(nucleotide for nucleotide in motifs[i] if nucleotide in nucleotidos)
```

```python
def ProfileMatrix(Motifs):

    # Orden: A C G T

    countMatrix = np.ones(shape = (4,len(Motifs[0])))

    for i in range(len(Motifs[0])):
        for j in range(len(Motifs)):

            if(Motifs[j][i] == 'A'):
                countMatrix[0][i] += 1
            elif(Motifs[j][i] == 'C'):
                countMatrix[1][i] += 1
            elif(Motifs[j][i] == 'G'):
                countMatrix[2][i] += 1
            else:
                countMatrix[3][i] += 1

    countMatrix = countMatrix/(sum(countMatrix)[0])

    return countMatrix
```

```python
def CountMatrix(Motifs):
    countMatrix = np.zeros(shape = (4,len(Motifs[0])))

    for i in range(len(Motifs[0])):
        for j in range(len(Motifs)):

            if(Motifs[j][i] == 'A'):
                countMatrix[0][i] += 1
            elif(Motifs[j][i] == 'C'):
                countMatrix[1][i] += 1
            elif(Motifs[j][i] == 'G'):
                countMatrix[2][i] += 1
            else:
                countMatrix[3][i] += 1
    
    return countMatrix
```

```python
def chain_probability(k_mer, profile):

    score = 1
    
    for i in range(len(k_mer)):
        if(k_mer[i] == 'A'):
            score *= profile[0][i]
        elif(k_mer[i] == 'C'):
            score *= profile[1][i]
        elif(k_mer[i] == 'G'):
            score *= profile[2][i]
        else:
            score *= profile[3][i]

    return score
```

```python
def obtain_most_probable_k_mer(profile, chain, k):
    best_chain = chain[0:k]
    for i in range(1, len(chain) - len(profile[0])):
        if(chain_probability(k_mer=chain[i:i+k], profile = profile) > chain_probability(k_mer=best_chain, profile = profile)):
            best_chain = chain[i:i+k]
    
    return best_chain
```

```python
def Score(motifs):
    dict_nucleotidos = {
        0: 'A',
        1 : 'C',
        2 : 'T',
        3: 'G'
    }

    countMatrix = CountMatrix(motifs)
    argmax = [np.argmax(countMatrix[:,j]) for j in range(len(countMatrix[0]))]
    popular_base = [dict_nucleotidos[j] for j in argmax]
    consensus = ''.join(nucleotide for nucleotide in popular_base)

    return int(sum([hamming(list(consensus), list(motifs[i]))*len(motifs[0]) for i in range(len(motifs))]))
```

```python
def RandomizedMotifSearch(Dna,k,epochs):
    n = len(Dna[0])
    random_positions = [np.random.randint(0,n-k) for i in range(len(Dna))]
    best_motifs = [Dna[i][random_positions[i]: random_positions[i]+k] for i in range(len(Dna))]

    for i in range(epochs):
        profile_matrix = ProfileMatrix(best_motifs)
        motifs = [obtain_most_probable_k_mer(profile=profile_matrix, chain = Dna[j], k = k) 
                  for j in range(len(Dna))]
        if(Score(motifs) <= Score(best_motifs)):
            best_motifs = motifs
            print(Score(best_motifs))
            print(best_motifs)
        
    return best_motifs
```
