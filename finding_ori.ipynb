### Implementación de primeros algoritmos para encontrar el Ori

Abro el archivo

```python
with open("../vibrio_cholerae_subset.txt") as f:
  chain = f.read()
```

Función frequency table

```python
def FrequencyTable(chain, k):
    freqMap = dict()
    n = len(chain)
    for i in range(n-k):
        Pattern = chain[i:i+k]
        if(Pattern not in freqMap.keys()):
            freqMap[Pattern] = 1
        else:
            freqMap[Pattern] += 1
    
    return freqMap
```

```python
diccionario = FrequencyTable(chain, 9)
diccionario
```

```python
maxKeys = lambda diccionario: [key for key in diccionario.keys() 
                if(diccionario[key] == diccionario[max(diccionario, key=diccionario.get)])]
```

```python
def BetterFrequentWords(chain, k):
    freqMap = FrequencyTable(chain, k)

    return maxKeys(freqMap)
    
```
