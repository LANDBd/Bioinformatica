En vez de pandas... podemos utilizar R estándar o dplyr

```r
# Base R
df <- data.frame(
  id = 1:5,
  grupo = c("A", "A", "B", "B", "C"),
  valor = c(10, 20, 30, 40, 50)
)

df2 <- data.frame(
  id = c(3, 4, 5, 6),
  categoria = c("X", "Y", "Z", "W")
)
```

```r
# Acceder a una columna (como vector)
df$valor

# Acceder a columna por índice
df[, 3]

# Acceder a fila por índice
df[2, ]

# Subdataframe por condición
df[df$valor > 25, ]

# Seleccionar múltiples columnas
df[, c("id", "valor")]
```

```r
library(dplyr)

df %>%
  group_by(grupo) %>%
  summarise(
    suma = sum(valor),
    promedio = mean(valor)
  )
```

```r
merge(df, df2, by = "id", all = FALSE)  # inner join

merge(df, df2, by = "id", all.x = TRUE)  # left join

merge(df, df2, by = "id", all.y = TRUE)  # right join

merge(df, df2, by = "id", all = TRUE)    # full join
```

```r
# dplyr: joins más legibles
left_join(df, df2, by = "id")

inner_join(df, df2, by = "id")

full_join(df, df2, by = "id")

anti_join(df, df2, by = "id")  #
```

```r
# Regresión lineal: mpg ~ wt (consumo vs peso del auto)
modelo_lm <- lm(mpg ~ wt, data = mtcars)
summary(modelo_lm)

# Graficar
plot(mtcars$wt, mtcars$mpg)
abline(modelo_lm, col = "red")
```

```r
# mpg en función de wt y hp (peso y caballos de fuerza)
modelo_lm2 <- lm(mpg ~ wt + hp, data = mtcars)
summary(modelo_lm2)
```
