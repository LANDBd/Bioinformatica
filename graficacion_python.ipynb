```python
from scipy import stats
```

```python
data = [1, 2, 2, 3, 4, 4, 4, 5, 6]

media = np.mean(data)
mediana = np.median(data)
moda = stats.mode(data, keepdims=False)
desviacion = np.std(data)

print("Media:", media)
print("Mediana:", mediana)
print("Moda:", moda.mode, " (Frecuencia:", moda.count, ")")
print("Desviación estándar:", desviacion)
```

```python
data = np.random.normal(loc=0, scale=1, size=50)

stat, p = stats.shapiro(data)

print("Estadístico:", stat)
print("p-valor:", p)

if p > 0.05:
    print("Parece distribución normal")
else:
    print("No parece distribución normal")
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Pearson
r_pearson, p_pearson = stats.pearsonr(x, y)
print("Pearson:", r_pearson, "p:", p_pearson)

# Spearman (rango)
r_spearman, p_spearman = stats.spearmanr(x, y)
print("Spearman:", r_spearman, "p:", p_spearman)
```

```python
# Semilla para reproducibilidad
np.random.seed(0)

# Generar datos base
n = 100
x1 = np.random.normal(0, 1, n)               # Variable base
x2 = x1 + np.random.normal(0, 0.2, n)        # Alta correlación con x1
x3 = -x1 + np.random.normal(0, 0.2, n)       # Correlación negativa con x1
x4 = 0.5 * x1 + 0.5 * x2 + np.random.normal(0, 0.3, n)  # Combinación de x1 y x2
x5 = np.random.normal(0, 1, n)               # Variable no correlacionada

# Stack de variables (matriz n x 5)
data = np.column_stack((x1, x2, x3, x4, x5))
labels = ['x1', 'x2', 'x3', 'x4', 'x5']

# Calcular matriz de correlación
corr_matrix = np.corrcoef(data, rowvar=False)

# Crear heatmap
fig, ax = plt.subplots()
cax = ax.imshow(corr_matrix, cmap='coolwarm', vmin=-1, vmax=1)

# Añadir etiquetas
ax.set_xticks(np.arange(len(labels)))
ax.set_yticks(np.arange(len(labels)))
ax.set_xticklabels(labels)
ax.set_yticklabels(labels)

# Rotar etiquetas del eje X
plt.setp(ax.get_xticklabels(), rotation=45, ha='right')

# Añadir valores numéricos sobre el mapa
for i in range(len(labels)):
    for j in range(len(labels)):
        ax.text(j, i, f"{corr_matrix[i, j]:.2f}", ha='center', va='center', color='black')

# Añadir barra de color
fig.colorbar(cax, label='Correlación')

plt.title('Heatmap de Correlación entre Variables')
plt.tight_layout()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D  # Necesario para gráficos 3D

# Semilla para reproducibilidad
np.random.seed(42)

# Datos simulados
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)
z = x + y + np.random.normal(0, 0.5, 100)  # z correlacionada con x + y

# Crear figura y ejes 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Graficar los puntos
sc = ax.scatter(x, y, z, c=z, cmap='viridis', marker='o')

# Etiquetas
ax.set_title('Gráfico 3D: Scatter Plot')
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')

# Barra de color para representar los valores de Z
fig.colorbar(sc, ax=ax, label='Valor de Z')

plt.tight_layout()
plt.show()
```

```python
import plotly.graph_objects as go

# Semilla para reproducibilidad
np.random.seed(42)

# Generar datos simulados
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)
z = x + y + np.random.normal(0, 0.5, 100)  # Correlación entre x, y y z

# Crear figura interactiva
fig = go.Figure(data=[go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size=5,
        color=z,                # Color por valor de z
        colorscale='Viridis',   # Escala de color
        opacity=0.8,
        colorbar=dict(title='Z')
    ),
    text=[f"X={xi:.2f}, Y={yi:.2f}, Z={zi:.2f}" for xi, yi, zi in zip(x, y, z)]  # Hover text
)])

# Configurar layout
fig.update_layout(
    title='Gráfico 3D Interactivo con Plotly',
    scene=dict(
        xaxis_title='Eje X',
        yaxis_title='Eje Y',
        zaxis_title='Eje Z'
    ),
    margin=dict(l=0, r=0, b=0, t=40)
)

# Mostrar figura (esto abre en navegador o inline si estás en Jupyter)
fig.show()
```

```python
from matplotlib.animation import FuncAnimation
from scipy.stats import norm, expon, uniform

# Configuración inicial
num_frames = 100
samples_per_frame = 100

# Distribuciones
distributions = {
    'Normal': {
        'rvs': lambda n: np.random.normal(loc=0, scale=1, size=n),
        'pdf': lambda x: norm.pdf(x, loc=0, scale=1),
        'range': (-5, 5)
    },
    'Exponencial': {
        'rvs': lambda n: np.random.exponential(scale=1, size=n),
        'pdf': lambda x: expon.pdf(x, scale=1),
        'range': (0, 10)
    },
    'Uniforme': {
        'rvs': lambda n: np.random.uniform(low=-2, high=2, size=n),
        'pdf': lambda x: uniform.pdf(x, loc=-2, scale=4),
        'range': (-3, 3)
    }
}

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
x_vals = {}
data = {key: [] for key in distributions}

# Precalcular x para PDFs
for i, (name, dist) in enumerate(distributions.items()):
    x_vals[name] = np.linspace(*dist['range'], 1000)

def update(frame):
    for ax in axes:
        ax.clear()

    for i, (name, dist) in enumerate(distributions.items()):
        # Generar nuevos datos
        new_data = dist['rvs'](samples_per_frame)
        data[name].extend(new_data)

        # Histogramas
        ax = axes[i]
        ax.hist(data[name], bins=80, density=True, alpha=0.6, label='Histograma')

        # PDF teórica
        x = x_vals[name]
        ax.plot(x, dist['pdf'](x), 'r-', lw=2, label='PDF teórica')

        # Estética
        ax.set_title(f"{name} - N = {len(data[name])}")
        ax.legend()
        ax.set_ylim(0, max(dist['pdf'](x)) * 1.5)

ani = FuncAnimation(fig, update, frames=num_frames, interval=200, repeat=False)
plt.tight_layout()
plt.show()
```
